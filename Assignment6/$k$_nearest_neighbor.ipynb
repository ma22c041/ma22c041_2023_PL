```python
import numpy as np
P = np.array([[0,0], [0.2, 0.22], [0.1, -0.1], [1,1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63], [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8,-0.23], [0.8, 0.6]])
k = 3
```

```python
import numpy as np
import matplotlib.pyplot as plt

def find_k_nearest_neighbors(points, k):

    # Calculate the pairwise distances between all points using numpy
    distances = np.linalg.norm(points[:, np.newaxis] - points, axis=2)

    # Set diagonal elements to a high value to exclude self as a neighbor
    np.fill_diagonal(distances, np.inf)

    # Find the indices of the k nearest neighbors for each point
    nearest_neighbors_indices = np.argsort(distances, axis=1)[:, :k]

    return nearest_neighbors_indices

# Find k nearest neighbors for each point
nearest_neighbors_indices = find_k_nearest_neighbors(P, k)

# Print the indices of k nearest neighbors for each point
for i, neighbors in enumerate(nearest_neighbors_indices):
    print(f"{i + 1}: {neighbors + 1}")

```

```python
# Plot the points
plt.scatter(P[:, 0], P[:, 1], c='blue', label='Points')

# Plot arrows from each point to its k-nearest neighbors
for i, neighbors in enumerate(nearest_neighbors_indices):
    for neighbor in neighbors:
        plt.arrow(P[i, 0], P[i, 1], P[neighbor, 0] - P[i, 0], P[neighbor, 1] - P[i, 1], head_width=0.05, head_length=0.01, fc='blue', ec='red', alpha=0.5)

# Add labels for the points
for i, point in enumerate(P):
    plt.text(point[0], point[1], f'P{i+1}', ha='right')

# Set axis limits and labels
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.title(f'K-Nearest Neighbors (k={k})')
plt.legend()
plt.grid()
plt.show()
```
