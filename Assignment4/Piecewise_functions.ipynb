```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000001)                                                 # Generate 1000001 uniformly spaced points in the interval [-5, 5].
y = np.where(x < 0, x * np.sin(x**2), np.cos(x))                                # Compute the values of the piecewise function using numpy vectorized operations.
                                                                                # For x < 0, calculate x * sin(x^2), and for x >= 0, calculate cos(x)
plt.figure(figsize=(6,5))                                                       # Create a figure for plotting with a specified size
plt.plot(x, y, label='Piecewise Function', color='orange')                      # Plot the piecewise function.
plt.xlabel('x')
plt.ylabel('f(x)')                                                              # Set the labels for the x-axis and y-axis.
plt.title('Piecewise Function Plot')                                            # Set the title of the plot
plt.grid(True)                                                                  #Add gridlines to the plot.
plt.legend()                                                                    # Display a legend for the plotted line.
plt.show()                                                                      # Show the plot.

```
