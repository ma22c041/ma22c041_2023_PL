```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)                    # Create an array of 100 evenly spaced points from 0 to 10.
y = np.sin(x)                                  # Calculate the sine of each point in the 'x' array.

plt.figure(figsize=(8, 4))                     # Create a new figure with a specified size (8 inches in width, 4 inches in height).
plt.plot(x, y, label='sin(x)', color='brown')  # Plot 'x' against 'y' with a brown line and add a label.
plt.xlabel('x')                                # Set the label for the x-axis.
plt.ylabel('sin(x)')                           # Set the label for the y-axis.
plt.title('Graph of sin(x) for x in [0, 10]')  # Set the title of the plot.
plt.grid(True)                                 # Add gridlines to the plot.
plt.show()                                     # Display the plot.
```
