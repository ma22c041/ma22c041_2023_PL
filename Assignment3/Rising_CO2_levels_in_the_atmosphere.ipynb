```python
import urllib.request

url = "ftp://aftp.cmdl.noaa.gov/products/trends/co2/co2_mm_mlo.txt"
data_file = "co2_mm_mlo.txt"

urllib.request.urlretrieve(url, data_file)

import numpy as np
data = np.genfromtxt(data_file, skip_header=72)  # Skip the header lines
data = data[~np.isnan(data[:, 3])]  # Remove rows with NaN values in the 'average' column

years = data[:, 0]
months = data[:, 1]
co2_levels = data[:, 3]

# Calculate the date in decimal years
decimal_years = years + months / 12.0

# Extract monthly averages
monthly_averages = co2_levels

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(decimal_years, monthly_averages, label="Monthly Averages")
plt.xlabel("Year")
plt.ylabel("CO2 Levels (ppm)")
plt.title("Monthly Averages of CO2 Levels")
plt.legend()
plt.grid(True)
plt.show()

# Load the data
data_file = "co2_mm_mlo.txt"
data = np.genfromtxt(data_file, skip_header=72)
data = data[~np.isnan(data[:, 3])]

# Calculate the date in decimal years
decimal_years = years + months / 12.0

# Calculate the time step (difference between consecutive time points)
time_step = decimal_years[1] - decimal_years[0]

# Calculate the rate of change using central difference (numerical differentiation)
co2_rate_of_change_calc = np.gradient(co2_levels, time_step)

# Plot the calculated rate of change
plt.figure(figsize=(10, 6))
plt.plot(decimal_years, co2_rate_of_change_calc, label="Calculated Rate of Change")
plt.xlabel("Year")
plt.ylabel("CO2 Rate of Change (ppm/year)")
plt.title("Calculated CO2 Rate of Change")
plt.legend()
plt.grid(True)
plt.show()
```
